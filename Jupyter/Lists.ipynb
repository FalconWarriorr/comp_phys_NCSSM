```python
# List Maniplation
# Coach Payne
# February, 2018
# Create a list, and use methods to manipulate the members of a list
# This is a compilation of various list manipulations

numbers = [1,2,5,78,24,35]
print ("The current list of numbers includes the following: \n")
print (numbers)
print ("\nThis also works with elements. The fifth element(numbers[4]) is \n")
print (numbers[4])
print ("Remember: the index numbers for each element begin at '0.")
print ("So numbers[0] is 1, numbers[3] is 78, and so on.\n")
print ("\nWe can even manipulate the amount of elements")
print (numbers * 2)
print ("\nNow examine the list to see if 'numbers' has changed:\n")
print (numbers)
print ("\nWe can also concatenate, or add a list to a list:\n")
numbers2 = [8,23,17,5]
print ("\nThis is a second list of numbers:\n")
print (numbers2)
numbers3 = numbers + numbers2
print ("\nAnd this is the concatenation of the two lists:\n")
print ("List 1:", numbers, "plus", numbers2, "is\n")
print (numbers3, ".")

print ("\nWe can also swap positions within a list.\n")
print ("The first list was: \n")
print (numbers)
print ("\nSwapping first and last: \n")

numbers[0], numbers[-1] = numbers[-1], numbers[0]
print (numbers)

print ("\nPut them in numerical order:\n")
numbers.sort()
print (numbers)
print("\nNow reverse the order.\n")
numbers.reverse()
print (numbers)

```

```python
# Tally occurrences of words in a list
# Through the Looking Glass is open source, and is downloadable from
# http: //www.gutenberg.org/ebooks/12
# Make sure to use a .txt version - download it and save as above into the working directory.

from collections import Counter

cnt = Counter()
for word in ['orange', 'apple', 'peach', 'apple', 'peach', 'banana', 'pear', 'peach', 'tangerine']:
    cnt[word] += 1
print ('The number of times each color appears in the list: ') 
print(cnt)
# prints Counter({'peach': 3, 'apple': 2, 'pear': 1, 'tangerine': 1, 'orange': 1, 'banana': 1}) 

# Another use for Counter()
# Find the ten most common words in Through the Looking Glass
# 're' below refers to 'regular expressions' which is actually a mini-programming
# sequence of instructions allowing the programmer to better process the material
# being analyzed - more at https://docs.python.org/2/howto/regex.html - you would
# still need to import Counter() from collections as above

import re
words = re.findall(r'\w+', open('looking_glass.txt').read().lower())
print ('\nThe top ten words used in "Through the Looking Glass": ')
print(Counter(words).most_common(10))
print ('Note: to simplify the search, all letters in all words were changed')
print ('      to lower case.')

#prints the following:
# [('the', 1775), ('and', 975), ('a', 819), ('to', 817), ('you', 686), ('it', 681),
# ('i', 642), ('of', 604), ('she', 544), ('said', 473)]

```
