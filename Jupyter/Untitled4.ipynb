```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array((0,0))
data = np.loadtxt("periodicity/star_wavelength.csv", str)
print(data)
data[0,0] = '1'
data = data.astype(float)
print(data)


plt.plot(data[:,0], data[:,1])
plt.show()
```

This data has no significant periodicity because it is just exponential growth.

```python
# import numpy as np
import matplotlib.pyplot as plt
data = np.array((0,0))
data = np.loadtxt("periodicity/stars.txt", float)
print(data)

c = np.fft.rfft(data[:,1])


plt.subplot(211)
plt.plot(data[:,0], data[:,1])

plt.subplot(212)
plt.plot(abs(c[50:-1])**2)
plt.xlim((0,200))
plt.show()
```

This data has a significant frequency at about 112 that we can see from the fourier transform.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("periodicity/stm.txt", float)


from mpl_toolkits.mplot3d import Axes3D
c = np.fft.rfft(data)


plt.imshow(data)

fig = plt.figure()
shape = data.shape
xs = np.arange(shape[0])
ys = np.arange(shape[1])
mesh = np.meshgrid(xs, ys)
print(mesh[0].shape)
print(mesh)
ax = fig.add_subplot(111, projection='3d')

ax.scatter(mesh[0], mesh[1], data)
```

```python
'''
=================
3D wireframe plot
=================

A very basic demonstration of a wireframe plot.
'''

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)

print(X)
print(Y)
# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
