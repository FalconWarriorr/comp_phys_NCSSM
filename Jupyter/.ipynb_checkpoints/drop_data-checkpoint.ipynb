```python
from numpy import loadtxt, polyfit, poly1d, arange
from pylab import plot, xlabel, ylabel, show

a = loadtxt("Drop_Data.csv", float, delimiter=',')
t1 = a[:,0]
t2 = []
t3 = []
y = a[:,1]
v = []
a = []
for i in range(len(y)-1):
    v.append((y[i+1]-y[i])/(t1[i+1]-t1[i]))
for i in range(len(t1)-1):
    t2.append((t1[i+1]+t1[i])/2)
for i in range(len(v)-1):
    a.append((v[i+1]-v[i])/(t2[i+1]-t2[i]))
for i in range(len(t2)-1):
    t3.append((t2[i+1]+t2[i])/2)    


coefficients1 = polyfit(t1,y,6)
polynomial1 = poly1d(coefficients1)
residuals_y = []
for i in range(len(y)):
    residuals_y.append(y[i] - polynomial1(t1[i]))
t1s = arange(0, max(t1)+0.1, 0.1)
ys = polynomial1(t1s)
plot(t1,y,'o')
plot(t1s,ys)
ylabel('y')
xlabel('t')
show()
plot(t1,residuals_y)
ylabel('y residuals')
xlabel('t')
show()

coefficients2 = polyfit(t2,v,6)
polynomial2 = poly1d(coefficients2)
residuals_v = []
for i in range(len(y)-1):
    residuals_v.append(v[i] - polynomial2(t2[i]))
t2s = arange(0, max(t2)+0.1, 0.1)
vs = polynomial2(t2s)
plot(t2,v,'o')
plot(t2s,vs)
ylabel('v')
xlabel('t')
show()
plot(t2,residuals_v)
ylabel('v residuals')
xlabel('t')
show()

coefficients3 = polyfit(t3,a,6)
polynomial3 = poly1d(coefficients3)
residuals_a = []
for i in range(len(y) - 2):
    residuals_a.append(a[i] - polynomial3(t3[i]))
t3s = arange(0, max(t3)+0.1, 0.1)
As = polynomial3(t3s)
plot(t3,a,'o')
plot(t3s,As)
ylabel('a')
xlabel('t')
show()
plot(t3,residuals_a)
ylabel('a residuals')
xlabel('t')
show()

```
