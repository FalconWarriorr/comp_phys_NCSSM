```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("world_records.csv", dtype = str, delimiter=',')
distance = []

time_m = []
time_f = []
time_avg = []

speed_m = []
speed_f = []
speed_avg = []

for i in data:
    #Create an array of the distances
    distance.append(float(i[0]))
    
    #Get the male time for each distance
    time_m_ = i[1].split(':')
    time_m_s = 0
    #Loop through an array possibly containing [hours, mins, sec]
    for k in range(len(time_m_), 0, -1):
        time_m_s += float(float(time_m_[k-1])*(60**(len(time_m_)-k)))
    
    #Get the female time for each distance
    time_f_ = i[2].split(':')
    time_f_s = 0
    #Loop through an array possibly containing [hours, mins, sec]    
    for k in range(len(time_f_), 0, -1):
        time_f_s += float(float(time_f_[k-1])*(60**(len(time_f_)-k)))
        
    #Add times to respective arrays
    time_m.append(time_m_s)
    time_f.append(time_f_s)
    time_avg.append((time_m_s+time_f_s)/2)
    
    #Determine average speed for each distance
    speed_f.append((distance[len(distance)-1])/(time_m[len(time_m)-1]))
    speed_m.append((distance[len(distance)-1])/(time_f[len(time_f)-1]))
    speed_avg.append((distance[len(distance)-1])/(time_avg[len(time_avg)-1]))
    

#PLot times per distance
plt.plot(distance, time_m)
plt.plot(distance, time_f)
plt.plot(distance, time_avg)
#Create the legend
plt.legend(["Male", "Female", "Avg"])
plt.grid()
plt.show()


plt.plot(distance, speed_m)
plt.plot(distance, speed_f)
plt.plot(distance, speed_avg)
#Create the legend
plt.legend(['Male', 'Female', 'Avg'])
plt.grid()
plt.show()
```
