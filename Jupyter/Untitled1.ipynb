```python
from numpy import linspace, pi, cos, sin
from matplotlib.pyplot import ion, figure, show
from mpl_toolkits.mplot3d import Axes3D

theta = linspace(0,2*pi,401)
a = 0.3
m = 11
n = 9
x = (1+a*cos(n*theta))*cos(m*theta)
y = (1+a*cos(n*theta))*sin(m*theta)
z = a*sin(n*theta)

ion()
fig = figure()
ax = Axes3D(fig)
ax.plot(x,y,z,'g',linewidth=4)
ax.set_zlim3d(-1.0, 1.0)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("A spiral as a parametric curve", weight="bold", size=16)
ax.elev, ax.azim = 60, -120
show()
```

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

time = mdates.drange(datetime.datetime(2018, 1, 1,),
                     datetime.datetime(2019, 1, 1,),
                     datetime.timedelta(days=10))
y1 = np.cumsum(np.random.random(time.size)-0.5)
y2 = np.cumsum(np.random.random(time.size)-0.5)

fig = plt.figure()

plt.plot_date(time, y1, 'b-')
plt.plot_date(time, y2, 'g-')

fig.autofmt_xdate()
plt.show()
```
