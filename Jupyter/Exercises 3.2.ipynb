## Deltoid curve plotting

```python
from pylab import show, plot
from math import sin, cos, pi

#Define resolution of polar curve
theta = 0
dtheta = 0.01

#Array for tracking x and y values
xlist = []
ylist = []
while theta < 2*pi:
    xlist.append(2*cos(theta) + cos(2*theta))
    ylist.append(2*sin(theta) - sin(2*theta))
    theta += dtheta
    

plot(xlist, ylist)
show()
```

## Galilean spiral

```python
from pylab import show, plot
from math import sin, cos, pi, exp

#Define resolution of polar curve
theta = 0
dtheta = 0.01

#Arrays for tracking graph data
rlist = []
xlist = []
ylist = []

#Parameters for the curves
a = 1
b = 2

#Define curve function
def spiral(theta):
    return (a + b*(theta**2))
#Convert polar to cartesian coordinates
def convert_to_cartesian(r, theta):
    return r*cos(theta), r*sin(theta)

#Generate the graph
while theta < 10*pi:
    r = spiral(theta**2)
    rlist.append(r)
    x, y = convert_to_cartesian(r, theta)
    xlist.append(x)
    ylist.append(y)
    theta += dtheta
    
#Plot the curve
plot(xlist, ylist)
show()
```

## Fey's function

```python
from pylab import show, plot
from math import sin, cos, pi, exp

#Define resolution of polar curve
theta = 0
dtheta = 0.01

#Arrays for tracking graph data
rlist = []
xlist = []
ylist = []

#Define curve function
def spiral(theta):
    return (exp(cos(theta)) - 2*cos(4*theta) + (sin(theta/12.0))**5)
#Convert polar to cartesian coordinates
def convert_to_cartesian(r, theta):
    return r*cos(theta), r*sin(theta)

#Generate the graph
while theta < 24*pi:
    r = spiral(theta)
    rlist.append(r)
    x, y = convert_to_cartesian(r, theta)
    xlist.append(x)
    ylist.append(y)
    theta += dtheta
    
#Plot the curve
plot(xlist, ylist)
show()
```
